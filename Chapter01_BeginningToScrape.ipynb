```python
from urllib.request import urlopen

html = urlopen('http://pythonscraping.com/pages/page1.html')
if not html.headers.get_content_charset() is None:
    print(html.read().decode(html.headers.get_content_charset()))
else:
    print(html.read())
```

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError

try:
#     html = urlopen("https://pythonscrapingthisurldoesnotexist.com")
    html = urlopen("http://detail.zol.com.cn/cell_phone/index1205447.shtml")
except HTTPError as e:
    print("The server returned an HTTP error")
except URLError as e:
    print("The server could not be found!")
else:
    if html.headers.get_content_charset() is None:
        print(html.read())
    else:
        print(html.read().decode(html.headers.get_content_charset()))
```

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup


def getTitle(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return None
    try:
        bsObj = BeautifulSoup(html.read(), "lxml")
        title = bsObj.body.h1
    except AttributeError as e:
        return None
    return title


title = getTitle("http://www.pythonscraping.com/pages/page1.html")
if title == None:
    print("Title could not be found")
else:
    print(title)
```
